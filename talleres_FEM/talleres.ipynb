```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
```

```python
x_1=np.linspace(1,5,5)
x_2=np.linspace(5,5,5)
x_3=np.linspace(1,1,5)
x_4=np.linspace(5,1,5)
```

```python
fig = plt.figure(figsize=(6,6))
spec = fig.add_gridspec(ncols=1,nrows=1, figure=fig)
ax1 = fig.add_subplot(spec[0,0])
ax1.spines[['right', 'top']].set_visible(False)
ax1.spines[["left", "bottom"]].set_position(("data", 3))
ax1.plot(1, 3, ">k", transform=ax1.get_yaxis_transform(), clip_on=False)
ax1.plot(3, 1, "^k", transform=ax1.get_xaxis_transform(), clip_on=False)
ax1.plot(x_1,x_3,'--',color='#d676ba')
ax1.plot(x_3,x_1,'--',color='#d676ba')
ax1.plot(x_2,x_1,'--',color='#d676ba')
ax1.plot(x_1,x_2,'--',color='#d676ba')
ax1.plot(x_1,x_1,'--',color='#d676ba')
ax1.plot(x_1,x_4,'--',color='#d676ba')
ax1.plot(5,5, 'o', color='#1f5070',markersize=12)
ax1.plot(1,1, 'o', color='#1f5070',markersize=12)
ax1.plot(1,5, 'o', color='#1f5070',markersize=12)
ax1.plot(5,1, 'o', color='#1f5070',markersize=12)
ax1.plot(3,3, 'o', color='#1f5070',markersize=12)
ax1.text(5,5.1,r'+ 2q',fontsize=12,style='italic',fontweight='bold')
ax1.text(0.68,5.11,r'+ q',fontsize=12,style='italic',fontweight='bold')
ax1.text(0.6,0.8,r'+ 2q',fontsize=12,style='italic',fontweight='bold')
ax1.text(5.1,0.8,r'- 4q',fontsize=12,style='italic',fontweight='bold')
ax1.text(3.1,3.1,r'+ q',fontsize=12,style='italic',fontweight='bold')
ax1.text(0.8,3.1,r'b',fontsize=12,style='italic',fontweight='bold')
ax1.text(3.1,0.8,r'b',fontsize=12,style='italic',fontweight='bold')
ax1.text(5.1,4.8,'(A)',style='italic',fontsize=11)
ax1.text(1.1,0.8,'(B)',style='italic',fontsize=11)
ax1.text(5.1,3.1,'x',style='italic')
ax1.text(3.1,5.1,'y',style='italic')
ax1.xaxis.set_major_locator(ticker.NullLocator())
ax1.yaxis.set_major_locator(ticker.NullLocator())
plt.savefig('Taller_FEM_fig_1', dpi = 300, bbox_inches = 'tight')
fig.show()
```
